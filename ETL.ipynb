```python
#imports needed for data etl process
import pandas as pd
import requests
from flask import jsonify
from pprint import pprint
```

Data Source Extraction

```python
#utilizes the nasa Global landslide catalog
url= 'https://data.nasa.gov/resource/dd9e-wu2v.json?$limit=50000'
#call the actual data
response= requests.get(url).json()
```

```python
#verify if response from json object was correct
print(response[0])
print(response[1])
print(response[11000])
```

```python
#initialize dataframe to hold values
landslide_basedf= pd.DataFrame(
#columns need: event_id,event_date,location accuracy, landslide category, landslide_trigger,landslide_size, landslide_setting, landslide injury count, country name, longitude,latitude    
columns=['landslide_id','landslide_date','located_accuracy','category_landslide',
'landslide_trig','landslide_area','landslide_env','injuries','fatalities',
'country','lng','lat'])
```

```python
#Collecting the rows of data needed, based on columns represented in landslide_basedf
for x in range(len(response)):
    #implement try and expect error for all missing values(nan) retrieved from dataset
    try:
        #landslide id
        event_id=response[x]['event_id']
        landslide_basedf.at[x,'landslide_id']=event_id
        #event date
        event_date=response[x]['event_date']
        landslide_basedf.at[x,'landslide_date']=event_date
        #location accuracy
        location_accuracy=response[x]['location_accuracy']
        landslide_basedf.at[x,'located_accuracy']=location_accuracy
        #landslide category
        landslide_category=response[x]['landslide_category']
        landslide_basedf.at[x,'category_landslide']=landslide_category
        #landslide trigger
        landslide_trigger=response[x]['landslide_trigger']
        landslide_basedf.at[x,'landslide_trig']=landslide_trigger
        #landslide size
        landslide_size=response[x]['landslide_size']
        landslide_basedf.at[x,'landslide_area']=landslide_size
        #landslide setting
        landslide_enviornment=response[x]['landslide_setting']
        landslide_basedf.at[x,'landslide_env']=landslide_enviornment
        #fatality count
        fatality_count=response[x]['fatality_count']
        landslide_basedf.at[x,'fatalities']=fatality_count
        #injury count
        injury_count=response[x]['injury_count']
        landslide_basedf.at[x,'injuries']=injury_count
        #country name
        country_name=response[x]['country_name']
        landslide_basedf.at[x,'country']=country_name
        #longitude
        longitude=response[x]['longitude']
        landslide_basedf.at[x,'lng']=longitude
        #latitude
        latitude=response[x]['latitude']
        landslide_basedf.at[x,'lat']=latitude
    except (KeyError):
        print(f'missing values present in row: {x}')
```

```python
#show dataframe
landslide_basedf
```

```python
#begin transforming data through identifying null values
landslide_basedf.isna().sum()
```

```python
# drop null values while still maintaing data integrity
transformed_ls_df=landslide_basedf.dropna(axis='index',how='any',subset=['located_accuracy', 'category_landslide', 'landslide_trig', 'landslide_area', 'landslide_env'])
```

```python
#drop columns with excessive missing data that can't be filled
transformed_ls_df=transformed_ls_df.drop(columns=['country', 'lng','lat'])
```

```python
#revalidate that correct values are presented, goal is to have no null values present 
transformed_ls_df.isna().sum()
```

```python
#replace null injuires and fatalities values with 0 as it is logical to assume that if no note of these counts  occured, then it didn't pose a presece to them
transformed_ls_df=transformed_ls_df.fillna(value='0')
```

```python
#revalidate that correct values are presented, goal is to have no null values present 
transformed_ls_df.isna().sum()
```

```python
#reset index to determine usable rows after null cleansing
transformed_ls_df=transformed_ls_df.reset_index(drop=True)
```

```python
#drop the time value in landslide_date based on empirical evidence that times are not truthful
transformed_ls_df['landslide_date'] = transformed_ls_df['landslide_date'].str.split('T').str[0] 
```
